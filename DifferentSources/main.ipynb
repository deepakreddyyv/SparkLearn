```python
options = {
    'header': 'true',
    'mode': 'FAILFAST',
    'path': '/home/deepak/programs/python/sparkLearn/data/flight-data/csv'
}

df = spark.read.format('csv').options(**options).load()

# permissive :- Sets all fields to null when it encounters a corrupted record and places all corrupted records in a string column called _corrupt_record
# dropMalformed :- Drops the row that contains malformed records
# failFast :- Fails immediately upon encountering malformed records
```

```python
# Write API
# DataFrameWriter.format(...).option(...).partitionBy(...).bucketB
# y(...).sortBy(
# ...).save()

# PartitionBy, bucketBy, and sortBy work only for file-based data
```

```python
df.write.format("PARQUET")\
.option("dateFormat", "yyyy-MM-dd")\
.option("path", "./save")\
.mode('OVERWRITE')\
.save()
```
